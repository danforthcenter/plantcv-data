```python
# Read hyperspectral data

# Path to leaf scan that we have been using in the hyperspectral documentation 
scandir = ""
imgname = "ee362e84-c861-4f0d-abbb-215ac58eed60_data"
filepath = os.path.join(scandir, imgname)

# Read in hyperspectral image 
array = pcv.readimage(filename=filepath, mode="envi")
```

```python
# load mask
file_mask = "mask.pkl"

mask = pkl.load(open(file_mask,"rb"))
pcv.plot_image(mask)
```

```python
# Showing histograms for 3 default wavelengths
pcv.params.debug = 'plot'
# pcv.params.debug = 'print'
hist_figure1 = pcv.visualize.hyper_histogram(array, mask)
pcv.print_image(hist_figure1, "histogram_default.png")
```

```python
# Showing the histogram for a single wavelength (700nm)
hist_figure2 = pcv.visualize.hyper_histogram(array, mask=mask, wvlengths=[700])
pcv.print_image(hist_figure2, "histogram_700.png")
```

```python
# Showing the histogram for two wavelengths (the 970nm is not in the range of visible spectrum)
hist_figure3 = pcv.visualize.hyper_histogram(array, mask=mask, wvlengths=[380, 970])
pcv.print_image(hist_figure3, "histogram_380_970.png")
```
